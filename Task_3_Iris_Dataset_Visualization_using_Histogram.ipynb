## **Plot the Histogram & Bar Chart for Numercal Data:**

###**Numerical Data-**



####**1. Sepal Length**

```python
# Histogram for Sepal Length
plt.figure(figsize=(10, 6))
sns.histplot(iris['sepal_length'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Sepal Length in Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot for Sepal Length
plt.figure(figsize=(12, 6))
sns.barplot(x='sepal_length', y='species', data=iris, palette='viridis')
plt.title('Bar Chart of Sepal Length by Species in Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Species')
plt.show()
```

#### **2. Sepal Width**

```python
# Histogram for Sepal Width
plt.figure(figsize=(10, 6))
sns.histplot(iris['sepal_width'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Sepal Width in Iris Dataset')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for Sepal Width
plt.figure(figsize=(10, 6))
sns.countplot(x='sepal_width', data=iris, palette='pastel')
plt.title('Distribution of Sepal Width in Iris Dataset')
plt.xlabel('Sepal Width')
plt.ylabel('Count')
plt.show()
```

```python
# Handling outliers
lower_bound = df['Fare'].quantile(0.25)
upper_bound = df['Fare'].quantile(0.75)
IQR = upper_bound - lower_bound

# Define a threshold for outliers
threshold = 1.5

# Identify outliers
outliers = df[(df['Fare'] < (lower_bound - threshold * IQR))
       | (df['Fare'] > (upper_bound + threshold * IQR))]

# Remove outliers
titanic_df_no_outliers = df[(df['Fare'] >= (lower_bound - threshold * IQR)) & (df['Fare'] <= (upper_bound + threshold * IQR))]
df
```

```python
# Display the cleaned dataset
print("Cleaned DataFrame:\n", df.head())
```
